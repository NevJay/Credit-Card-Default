# 3. DATA CLEANING, STANDARDIZING AND TRAIN & TEST SPLIT

```python
credit_df = credit_df.drop_duplicates()
print(credit_df.duplicated().sum())
```

```python
#Feature Creation
credit_df["AVG_BILL_AMT"] = (credit_df["BILL_AMT1"] + credit_df["BILL_AMT2"] + credit_df["BILL_AMT3"] + credit_df["BILL_AMT4"] + credit_df["BILL_AMT5"] + credit_df["BILL_AMT6"])/6
credit_df["AVG_PAY_AMT"] = (credit_df["PAY_AMT1"] + credit_df["PAY_AMT2"] + credit_df["PAY_AMT3"] + credit_df["PAY_AMT4"] + credit_df["PAY_AMT5"] + credit_df["PAY_AMT6"])/6 
credit_df.drop(columns=['BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6', 'PAY_AMT1', 'PAY_AMT2', 'PAY_AMT3', 'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6'], axis=1, inplace=True)
```

```python
#creating target values
y = credit_df['default.payment.next.month']
y
```

```python
# one hot encoding for 'SEX', 'MARRIAGE' & 'EDUCATION'
X = pd.get_dummies(data=credit_df, columns=['SEX', 'MARRIAGE', 'EDUCATION'])
X.drop(columns=['default.payment.next.month'], axis=1, inplace=True)
X.columns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, stratify = y, random_state=42)
X_train, X_valid, y_train, y_valid = train_test_split(X_train, y_train, test_size = 0.2, stratify = y_train, random_state=42)
print("Variables in Train Set : {} & Test Set : {} & Validation Set : {}".format(len(X_train), len(X_test), len(X_valid)))

#scaler = StandardScaler()
ct = ColumnTransformer([
        ('somename', StandardScaler(), ['LIMIT_BAL', 'AGE', 'AVG_BILL_AMT', 'AVG_PAY_AMT'])
    ], remainder='passthrough')
X_train_scaled = ct.fit_transform(X_train)
X_test_scaled = ct.transform(X_test)
X_valid_scaled = ct.transform(X_valid)
```

# 4: FEATURE SELECTION USING XGBOOST

```python
import xgboost as xgb

model = xgb.XGBClassifier(objective='binary:logistic', learning_rate = 0.1, max_depth = 5, n_estimators = 100)
model.fit(X_train_scaled, y_train, eval_set= [(X_train_scaled, y_train), (X_valid_scaled, y_valid)])
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

y_pred = model.predict(X_test_scaled)
print("Accuracy {:.2f} %".format( 100 * accuracy_score(y_pred, y_test)))
```

```python
# Testing Set Performance
cm = confusion_matrix(y_pred, y_test)
print(cm)
sns.heatmap(cm, annot=True);
print("Classification Report :- \n",classification_report(y_test, y_pred))
```

```python
fig, (ax) = plt.subplots(ncols=1, figsize=(20,15))
xgb.plot_importance(model, height=0.8, title="Features importance (XGBoost)", ax=ax, color="green") 
plt.show()
```

# 5: TRAINING AND EVALUATION OF THE MODEL

```python
#creating target values
y = credit_df['default.payment.next.month']
y
```

```python
X = pd.get_dummies(data=credit_df, columns=['SEX', 'MARRIAGE', 'EDUCATION'])
X.drop(columns=['EDUCATION_0', 'EDUCATION_1', 'EDUCATION_5', 'EDUCATION_2', 'MARRIAGE_1', 'SEX_2', 'EDUCATION_3', 'MARRIAGE_2', 'MARRIAGE_0', 'EDUCATION_4', 'MARRIAGE_3', 'default.payment.next.month'], axis=1, inplace=True)
X.columns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, stratify = y, random_state=42)
X_train, X_valid, y_train, y_valid = train_test_split(X_train, y_train, test_size = 0.2, stratify = y_train, random_state=69)
print("Variables in Train Set : {} & Test Set : {} & Validation Set : {}".format(len(X_train), len(X_test), len(X_valid)))

print(y_train.value_counts())
#scaler = StandardScaler()
ct = ColumnTransformer([
        ('somename', StandardScaler(), ['LIMIT_BAL', 'AGE', 'AVG_BILL_AMT', 'AVG_PAY_AMT'])
    ], remainder='passthrough')

X_train_scaled = ct.fit_transform(X_train)
X_test_scaled = ct.transform(X_test)
X_valid_scaled = ct.transform(X_valid)
```

```python
import xgboost as xgb

model = xgb.XGBClassifier(objective='binary:logistic', learning_rate = 0.1, max_depth = 5, n_estimators = 100, random_state = 42)
model.fit(X_train_scaled, y_train, eval_set= [(X_train_scaled, y_train), (X_valid_scaled, y_valid)])
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

y_pred = model.predict(X_test_scaled)
print("Accuracy {:.2f} %".format( 100 * accuracy_score(y_pred, y_test)))
```

```python
# Testing Set Performance
cm = confusion_matrix(y_pred, y_test)
print(cm)
sns.heatmap(cm, annot=True);
print("Classification Report :- \n",classification_report(y_test, y_pred))
```

```python
fig, (ax) = plt.subplots(ncols=1, figsize=(15,15))
xgb.plot_importance(model, height=0.8, title="Features importance (XGBoost)", ax=ax, color="green") 
plt.show()
```

# 6: HYPERPARAMETER TUNING & EVALUATION OF BEST ESTIMATOR

```python
from sklearn.model_selection import RandomizedSearchCV

param_grid = {'gamma': [1, 5],   # regularization parameter
              'subsample': [0.8, 1.0], # % of rows taken to build each tree
              'colsample_bytree': [0.8, 1.0], # number of columns used by each tree
              'max_depth': [7, 5], # depth of each tree
              'n_estimators': [100, 300] #no. of trees
             }

xgb_model = xgb.XGBClassifier(objective='binary:logistic', learning_rate = 0.01)
grid = RandomizedSearchCV(xgb_model, 
                    param_grid, 
                    refit = True, 
                    verbose = 4,
                    n_iter = 2
                   )
grid.fit(X_train_scaled, y_train)
```

```python
print("Best Score: {:.2f} %".format(grid.best_score_ * 100))
print("Best Parameters:", grid.best_params_)
model = grid.best_estimator_
```

```python
pred = model.predict(X_valid_scaled)

print("Accuracy {:.2f} %".format( 100 * accuracy_score(y_valid, pred)))
# Testing Set Performance
cm = confusion_matrix(y_valid, pred)
print(cm)
sns.heatmap(cm, annot=True);
print("Classification Report :- \n",classification_report(y_valid, pred))
```

```python
pred = model.predict(X_test_scaled)

print("Accuracy {:.2f} %".format( 100 * accuracy_score(y_test, pred)))
# Testing Set Performance
cm = confusion_matrix(y_test, pred)
print(cm)
sns.heatmap(cm, annot=True);
print("Classification Report :- \n",classification_report(y_test, pred))
```
